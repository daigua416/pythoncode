```python
import networkx as nx
import matplotlib.pyplot as plt

# 设置随机种子，以便结果可重复
seed = 42
# 节点数量
num_nodes = 1000
# 平均度
average_degree = 4

# 生成ER网络
er_graph = nx.erdos_renyi_graph(num_nodes, p=average_degree/(num_nodes-1), seed=seed)

# 打印网络信息
print("节点数量:", len(er_graph.nodes()))
print("边数量:", len(er_graph.edges()))
print("平均度:", sum(dict(er_graph.degree()).values()) / len(er_graph))

# 可视化网络
nx.draw(er_graph, with_labels=False, node_size=10)
plt.show()
```
